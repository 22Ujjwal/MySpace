# Plotting histogram

```python
mydf1[['CRVUSDT-open', 'CRVUSDT-close']].plot(title='Market Legend', figsize=(20, 10))
```

```python
import matplotlib.pyplot as plt
test_colors = mydf1['Movement'].apply(lambda x: 'red' if x < 0 else 'green').tolist()

plt.figure(figsize=(10, 4))
plt.bar(
    x=np.arange(len(mydf1)),
    height=mydf1['Movement'],
    color=test_colors
)
plt.title('CRVUSDT Movement')
plt.xlabel('TIME')
plt.ylabel('MOTION')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```
